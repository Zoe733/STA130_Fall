### Part a

A Classification Decision Tree is a supervised machine learning model that categorizes data points into specific classes or categories according to their features. This tree models decision-making by sequentially splitting the data based on the attributes or features that provide the most information at each node, eventually leading to leaf nodes where the final classification decisions are made.Issues Solved by Classification Decision Trees: Binary Classification: This addresses scenarios where each data point must be sorted into one of two categories, like deciding whether an email is spam or not. Multi-class Classification: This deals with situations where data points must be categorized into several different classes, such as distinguishing among various types of fruits in an image.Functionality of a Classification Decision Tree: Root Node: This node encompasses the whole dataset and initiates the data splitting according to the feature with the maximum information gain. Decision Nodes: These are points within the tree where further divisions of the data occur. Leaf Nodes: These are the final nodes that deliver the predictive results, determined by the path from the root node through the decision nodes.Real world applications:these trees are utilized in various fields, such as medical diagnosis to categorize patients based on symptoms and test results, in banking to assess loan risk levels, in retail for customer segmentation tailored to boost marketing efforts, in fraud detection by analyzing transaction patterns, and in manufacturing for fault diagnosis in equipment. 

### Part b

Classification Decision Tree

Purpose: Primarily employed for categorizing data into specific categories or classes, which are predefined.
Working Mechanism: The model takes the form of a tree with decision nodes and leaf nodes. Each decision node identifies a particular dataset feature, dividing the data according to specific criteria. The divisions are strategically made to enhance information gain at every stage, thereby creating more uniform groups in relation to the target variable. The route from the root node to a leaf node delineates a sequence of rule-based decisions that culminate in a classification decision. Prediction: When new data is inputted, it traverses through the tree following these established rules until it arrives at a leaf node. There, it is assigned a class label, typically based on the majority class or another standard among the training examples that gathered at that leaf.

Multiple linear regression 
Working Mechanism:
It assumes a linear relationship between the independent variables (features) and the dependent variable (target).
The model generates a regression equation of the form $$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n + \epsilon
$$Coefficients are determined during training by minimizing the sum of the squared differences between the observed and predicted values (least squares method).


Key Differences: Output Type: Decision trees generate categorical labels, whereas multiple linear regression produces continuous outcomes. Data Handling: Decision trees accommodate non-linear relationships and categorical features without specific data distribution assumptions, unlike multiple linear regression which requires a linear relationship and numeric inputs. Interpretability: Decision trees are easily traced to understand predictions due to their rule-based structure, making them clearer than linear regression models, which explain feature impacts via coefficients but may be less intuitive. Sensitivity to Outliers: Decision trees are less sensitive to outliers due to their data-segmenting structure, while outliers can significantly alter linear regression results.

### Distinctions:

Accuracy vs. Precision: Accuracy measures the model's correct predictions, both positive and negative, while precision focuses solely on the accuracy of positive predictions. Sensitivity vs. Specificity: These metrics assess model accuracy from the perspective of positive cases (sensitivity) and negative cases (specificity), useful in contexts where different errors carry different risks. Sensitivity vs. Precision: Sensitivity evaluates the percentage of true positives detected, while precision gauges the accuracy of these positive detections, crucial for situations where confirming true positives is as important as detecting them.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Load the data
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Data cleaning and preprocessing
ab.drop(['Weight_oz', 'Width', 'Height'], axis=1, inplace=True)
ab.dropna(inplace=True)
ab['Pub year'] = ab['Pub year'].astype(int)
ab['NumPages'] = ab['NumPages'].astype(int)
ab['Hard_or_Paper'] = ab['Hard_or_Paper'].astype('category')

# Descriptive statistics and visualization
print("Descriptive Statistics for numerical columns:")
print(ab.describe())
print("\nDistribution of Hard_or_Paper categories:")
print(ab['Hard_or_Paper'].value_counts())

plt.figure(figsize=(12, 6))
ab.boxplot(column=['List Price', 'Amazon Price', 'NumPages', 'Pub year'])
plt.title('Boxplot of Numerical Columns')
plt.show()

ab.hist(column=['List Price', 'Amazon Price', 'NumPages', 'Pub year'], bins=15, figsize=(12, 10))
plt.suptitle('Histograms of Numerical Columns')
plt.show()

# Splitting the dataset into training and testing sets with a random seed
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab, test_size=0.2, random_state=42)  # random_state set to 42 for reproducibility

# Reporting the number of observations in the training and testing datasets
print(f"Number of observations in the training dataset: {ab_reduced_noNaN_train.shape[0]}")
print(f"Number of observations in the testing dataset: {ab_reduced_noNaN_test.shape[0]}")
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Preparing the features and target based on the training dataset
X_train = ab_reduced_noNaN_train[['List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train['Hard_or_Paper'])['H']

X_test = ab_reduced_noNaN_test[['List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test['Hard_or_Paper'])['H']

# Initializing and training the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=2)  # Limiting the depth of the tree to 2
clf.fit(X_train, y_train)

# Plotting the decision tree
plt.figure(figsize=(16, 10))
plot_tree(clf, filled=True, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'])
plt.title('Decision Tree for Predicting Book Cover Type Based on List Price')
plt.show()

# Optionally, evaluate the model
from sklearn.metrics import accuracy_score, classification_report
y_pred = clf.predict(X_test)
print("Accuracy of the model:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Assuming ab_reduced_noNaN_train and ab_reduced_noNaN_test have been properly defined and are available in the environment
# Preparing the features and target based on the training dataset
X_train = ab_reduced_noNaN_train[['List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train['Hard_or_Paper'])['H']
X_test = ab_reduced_noNaN_test[['List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test['Hard_or_Paper'])['H']

# Initializing and training the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=2)  # Limiting the depth of the tree to 2
clf.fit(X_train, y_train)

# Plotting the decision tree
plt.figure(figsize=(16, 10))
plot_tree(clf, filled=True, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'])
plt.title('Decision Tree for Predicting Book Cover Type Based on List Price')
plt.show()

# Predicting the test set results
y_pred = clf.predict(X_test)

# Computing confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Calculating sensitivity, specificity, and accuracy
TP = cm[1, 1]  # True Positive: correctly predicted event values
TN = cm[0, 0]  # True Negative: correctly predicted no-event values
FP = cm[0, 1]  # False Positive: incorrectly predicted event values (Type I error)
FN = cm[1, 0]  # False Negative: incorrectly predicted no-event values (Type II error)

sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
accuracy = (TP + TN) / (TP + TN + FP + FN)

print(f"Sensitivity: {sensitivity:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"Accuracy: {accuracy:.2f}")

# Optionally, printing a detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
